## Exercise 5 - Regex

In the regex task section, syntax used for the example text was retrieved from here: https://loremipsum.io/.

```python
example_text = "Lorem ipsum dolorolo at sit amet, consectetur ipsumm adipiscing elit, sed do eiusmod tempor incididunt ut labore et dolore magna aliqua. Ut enim ad minim veniam, quis nostrud exercitation ullamco (laboris) nisi ut (aliquip) ex ea commodo consequat. [192.168.1.1:8080] Duis aute irure dolor in reprehenderit in voluptate velit esse cillum dolore eu fugiat nulla pariatur. Excepteur sint occaecat cupidatat non proident, sunt in culpa qui officia deserunt mollit anim id est laborum. [192.168.1.254:80]"
```

1 Add the necessary libraries.

```python
import re
```

2 Find how many times the word `ipsum` appears in the text.

```python
# Find how many times the *word* 'ipsum' appears in the text
word_count = len(re.findall(r"\bipsum\b", example_text))
print("Word 'ipsum' appears:", word_count, "times")
```

3 Find in the text how many times the string `ipsum` (not the word!) appears.

```python
# Find how many times the *string* 'ipsum' (not necessarily as a word) appears
string_count = len(re.findall(r"ipsum", example_text))
print("String 'ipsum' appears:", string_count, "times")
```

4 Search the text for everything after the word `Excepteur` (Note: the word Excepteur should not be included in the output!)

```python
# Find everything that appears *after* the word 'Excepteur' (excluding the word itself)
after_excepteur = re.search(r"Excepteur\s+(.*)", example_text)
if after_excepteur:
    print("Text after 'Excepteur':", after_excepteur.group(1))
```

5 Find all the words inside the brackets `()`.

```python
# Find all the words inside parentheses ()
inside_parentheses = re.findall(r"\((.*?)\)", example_text)
print("Words inside parentheses:", inside_parentheses)
```

6 Choose all the words that end with `at` (include the word 'at' also!).

```python
# Choose all the words that end with 'at' (include the word 'at' itself)
ends_with_at = re.findall(r"\b\w*at\b", example_text)
print("Words ending with 'at':", ends_with_at)
```

7 Find words that contain the string `olo` but do not start or end with that string.

```python
# Find words that contain the string 'olo' but do not start or end with it
contains_olo = re.findall(r"\b\w+olo\w+\b", example_text)
print("Words containing 'olo' but not starting/ending with it:", contains_olo)
```

8 Next find all IP addresses with the following form: `ip-address:port-number` (eg 192.168.1.1:8080) without static search or `r'\[.*?\]'`/similar syntax which returns all the content inside the square brackets.

```python
# Find all IP addresses in the form ip:port (e.g. 192.168.1.1:8080)
# without simply matching inside square brackets
ip_port = re.findall(r"\b\d{1,3}(?:\.\d{1,3}){3}:\d{1,5}\b", example_text)
print("IP addresses with ports:", ip_port)
```
