```python
# E1

import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
earthquake_data = pd.read_csv("data_files/earthquakes.csv")

# Convert the 'Date' column to datetime with the appropriate format
earthquake_data['Date'] = pd.to_datetime(earthquake_data['Date'], format='mixed')

# Filter earthquakes with magnitude 8.2 or over
high_magnitude_earthquakes = earthquake_data[earthquake_data['Magnitude'] >= 8.2]

# Create the bar chart
#plt.bar(high_magnitude_earthquakes['Date'], high_magnitude_earthquakes['Magnitude'], width=1)

# Sample a subset of earthquakes (adjust sample size as needed)
sampled_data = high_magnitude_earthquakes.sample(20)

plt.bar(sampled_data['Date'], sampled_data['Magnitude'], width=25)
# Add labels and title
plt.xlabel('Date')
plt.ylabel('Magnitude')
plt.title('Earthquakes with Magnitude 8.2 or Over')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, locator=plt.Date.YearLocator())

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import YearLocator, DateFormatter

# Read the CSV file into a DataFrame
earthquake_data = pd.read_csv("data_files/earthquakes.csv")

# Convert the 'Date' column to datetime with the appropriate format
earthquake_data['Date'] = pd.to_datetime(earthquake_data['Date'], format='mixed')

# Filter earthquakes with magnitude 8.2 or over
high_magnitude_earthquakes = earthquake_data[earthquake_data['Magnitude'] >= 8.2]

# Create the bar chart
plt.figure(figsize=(15, 6))  # Increase figure size for better readability
plt.bar(high_magnitude_earthquakes['Date'], high_magnitude_earthquakes['Magnitude'], width=21)

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Magnitude')
plt.title('Earthquakes with Magnitude 8.2 or Over')

# Configure x-axis ticks and labels
plt.gca().xaxis.set_major_locator(YearLocator())  # Set major ticks for every year
plt.gca().xaxis.set_major_formatter(DateFormatter('%Y'))  # Format labels as YYYY

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
earthquake_data = pd.read_csv("data_files/earthquakes.csv")

# Convert the 'Date' column to datetime with the appropriate format
earthquake_data['Date'] = pd.to_datetime(earthquake_data['Date'], format='mixed')

# Filter earthquakes with magnitude 8.2 or over
high_magnitude_earthquakes = earthquake_data[earthquake_data['Magnitude'] >= 8.2]

# Count the number of earthquakes per day
earthquakes_per_day = high_magnitude_earthquakes.groupby(high_magnitude_earthquakes['Date'].dt.date).size()

# Create the bar chart
plt.bar(earthquakes_per_day.index, earthquakes_per_day)

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Number of Earthquakes')
plt.title('Earthquakes with Magnitude 8.2 or Over (Number per Day)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import re

# Read the CSV file into a DataFrame
earthquake_data = pd.read_csv("data_files/earthquakes.csv")

# Replace non-standard characters in the 'Date' column with a consistent format
earthquake_data['Date'] = earthquake_data['Date'].str.replace('T', ' ').str.replace('Z', '')

# Extract the date portion using regular expressions
earthquake_data['Date'] = earthquake_data['Date'].apply(lambda x: re.search(r'\d{4}-\d{2}-\d{2}', x).group(0) if isinstance(x, str) and re.search(r'\d{4}-\d{2}-\d{2}', x) else pd.NaT)

# Convert the 'Date' column to datetime
earthquake_data['Date'] = pd.to_datetime(earthquake_data['Date'])

# Print the data type of the 'Date' column
print(earthquake_data['Date'].dtype)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import YearLocator, DateFormatter
import re

# Read the CSV file into a DataFrame
earthquake_data = pd.read_csv("data_files/earthquakes.csv")

# Convert the 'Date' column to datetime with the appropriate format
earthquake_data['Date'] = pd.to_datetime(earthquake_data['Date'], format='mixed')

# Filter earthquakes with magnitude 8.2 or over
high_magnitude_earthquakes = earthquake_data[earthquake_data['Magnitude'] >= 8.2]

# Create the bar chart
plt.figure(figsize=(15, 6))  # Increase figure size for better readability
plt.bar(high_magnitude_earthquakes['Date'], high_magnitude_earthquakes['Magnitude'], width=21)

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Magnitude')
plt.title('Earthquakes with Magnitude 8.2 or Over')

# Configure x-axis ticks and labels
plt.gca().xaxis.set_major_locator(YearLocator())  # Set major ticks for every year
plt.gca().xaxis.set_major_formatter(DateFormatter('%Y'))  # Format labels as YYYY

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()

# Replace non-standard characters in the 'Date' column with a consistent format
earthquake_data['Date'] = earthquake_data['Date'].astype(str).str.replace('T', ' ').str.replace('Z', '')

# Extract the date portion using regular expressions
earthquake_data['Date'] = earthquake_data['Date'].apply(lambda x: re.search(r'\d{4}-\d{2}-\d{2}', x).group(0) if isinstance(x, str) and re.search(r'\d{4}-\d{2}-\d{2}', x) else pd.NaT)

# Convert the 'Date' column to datetime
earthquake_data['Date'] = pd.to_datetime(earthquake_data['Date'])
```
