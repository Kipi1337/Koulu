```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the dataset
data = pd.read_csv("data_files/voice.csv")

# Select features and target variable
X = data[['meanfun', 'IQR']]
y = data['label']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=13)

# Define possible values of k to try
k_values = range(1, 21)  # Example: Try k values from 1 to 20

# Initialize lists to store accuracy scores for each value of k
accuracy_scores = []

# Iterate over each value of k
for k in k_values:
    # Create KNN classifier with current value of k
    knn = KNeighborsClassifier(n_neighbors=k)
    
    # Train the classifier
    knn.fit(X_train, y_train)
    
    # Make predictions on the testing set
    y_pred = knn.predict(X_test)
    
    # Calculate accuracy and store it
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

# Find the optimal value of k with the highest accuracy
optimal_k = k_values[accuracy_scores.index(max(accuracy_scores))]

print("Optimal value of k:", optimal_k)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# Load the dataset
data = pd.read_csv("data_files/voice.csv")

# Select features and target variable
X = data[['meanfun', 'IQR']]
y = data['label']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=13)

# Define possible values of k to try
k_values = range(1, 21)  # Example: Try k values from 1 to 20

# Initialize lists to store accuracy scores for each value of k
accuracy_scores = []

# Iterate over each value of k
for k in k_values:
    # Create KNN classifier with current value of k
    knn = KNeighborsClassifier(n_neighbors=k)
    
    # Train the classifier
    knn.fit(X_train, y_train)
    
    # Make predictions on the testing set
    y_pred = knn.predict(X_test)
    
    # Calculate accuracy and store it
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

# Find the optimal value of k with the highest accuracy
optimal_k = k_values[accuracy_scores.index(max(accuracy_scores))]

print("Optimal value of k:", optimal_k)

def classify_and_visualize(X, y, k):
    # Convert DataFrame to NumPy array
    if isinstance(X, pd.DataFrame):
        X = X.values
    
    # Check if X has at least two columns
    if X.shape[1] < 2:
        print("Error: X must have at least two columns.")
        return
    
    # Handle invalid or non-finite values
    X = np.nan_to_num(X)
    
    # Define classifier
    classifier = KNeighborsClassifier(n_neighbors=k)
    
    # Split data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
    
    # Fit the data for the classifier
    classifier.fit(X_train, y_train)
    
    # Calculate minimum and maximum values for boundaries for meanfun and IQR
    min_meanfun = np.min(X[:, 0])
    max_meanfun = np.max(X[:, 0])
    min_IQR = np.min(X[:, 1])
    max_IQR = np.max(X[:, 1])
    
    # Generate grid points for prediction
    xx, yy = np.meshgrid(np.linspace(min_meanfun, max_meanfun, 100),
                         np.linspace(min_IQR, max_IQR, 100))
    
    # Run predictions
    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    
    # Convert string labels to numerical values
    label_encoder = LabelEncoder()
    Z = label_encoder.fit_transform(Z)
    
    Z = Z.reshape(xx.shape)
    
    # Draw a chart from predicted data utilizing previously defined colormaps
    plt.figure(figsize=(10, 6))
    plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
    
    # Encode the string labels into numerical values for plotting
    y_train_encoded = label_encoder.transform(y_train)
    
    # Plot the training points with the encoded labels
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train_encoded, cmap=plt.cm.Paired, edgecolors='k')
    plt.xlabel('meanfun')
    plt.ylabel('IQR')
    plt.title(f'KNN Classification (k={k}), Accuracy: {accuracy_score(y_test, classifier.predict(X_test)):.2f}')
    plt.show()

# Call the function
classify_and_visualize(X, y, optimal_k)
```
