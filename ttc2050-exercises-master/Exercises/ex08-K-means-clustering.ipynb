## Exercise 8 - K-means clustering

- Download the dataset from [here](http://student.labranet.jamk.fi/~hantt/data/voice.csv)
- This dataset will be used in exercises 8, 9 and 10
- Dataset includes male and female voice samples with different features
- In this exercise you will use unsupervised machine learning method called K-means clustering for the dataset

1 Import all necessary libraries. Then load the dataset and save it to the dataframe. In this exercise you will use only meanfun (*average of fundamental frequency measured across acoustic signal*) and minfun (*minimum fundamental frequency measured across the acoustic signal*) columns. Filter data so that only forementioned columns will be included.

2 Run the predictions for the data so that it will be categorised for male and female data.

3 Visualise the data so that clusters for male and female data will be presented in scatter plot (centroids should also be included).

4 Validate K-means clustering prediction results using confusion matrix.

```python
# E1

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the dataset
data = pd.read_csv("data_files/voice.csv")

# Select columns
features = data[['meanfun', 'minfun']]

# Kmeans
kmeans = KMeans(n_clusters=1)

# Fit the model to the data
kmeans.fit(features)

plt.scatter(features['meanfun'], features['minfun'], c=kmeans.labels_, cmap='viridis')
plt.xlabel('meanfun')
plt.ylabel('minfun')
plt.title('K-means Clustering')
plt.legend()
plt.show()
```

```python
# E2

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the dataset
data = pd.read_csv("data_files/voice.csv")

# Select the columns
features = data[['meanfun', 'minfun']]
gender = data['label']

# Separate genders
male_data = features[gender == 'male']
female_data = features[gender == 'female']

# Fit Kmeans models to the data
kmeans_male = KMeans(n_clusters=1, random_state=13)
kmeans_male.fit(male_data)
kmeans_female = KMeans(n_clusters=1, random_state=13)
kmeans_female.fit(female_data)

# Predict the future data
predictions_male = kmeans_male.predict(male_data)
predictions_female = kmeans_female.predict(female_data)

# Visualize male data
plt.scatter(male_data['meanfun'], male_data['minfun'], c=predictions_male, cmap='coolwarm', label='Male Data')


# Visualize female data
plt.scatter(female_data['meanfun'], female_data['minfun'], c=predictions_female, cmap='viridis', label='Female Data')


# Add labels and title
plt.xlabel('meanfun')
plt.ylabel('minfun')
plt.title('K-means Clustering for Male and Female Data')
plt.legend()
plt.show()
```

```python
# E3

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load the dataset
data = pd.read_csv("data_files/voice.csv")

# Select the columns
features = data[['meanfun', 'minfun']]
gender = data['label']

# Separate genders
male_data = features[gender == 'male']
female_data = features[gender == 'female']

# Fit Kmeans models to the data
kmeans_male = KMeans(n_clusters=1, random_state=13)
kmeans_male.fit(male_data)
kmeans_female = KMeans(n_clusters=1, random_state=13)
kmeans_female.fit(female_data)

# Predict the future data
predictions_male = kmeans_male.predict(male_data)
predictions_female = kmeans_female.predict(female_data)

# Visualize male data
plt.scatter(male_data['meanfun'], male_data['minfun'], c=predictions_male, cmap='coolwarm', label='Male Data')
plt.scatter(kmeans_male.cluster_centers_[:, 0], kmeans_male.cluster_centers_[:, 1], marker='x', color='red', label='Male Centroids')

# Visualize female data
plt.scatter(female_data['meanfun'], female_data['minfun'], c=predictions_female, cmap='viridis', label='Female Data')
plt.scatter(kmeans_female.cluster_centers_[:, 0], kmeans_female.cluster_centers_[:, 1], marker='x', color='blue', label='Female Centroids')

# Add labels and title
plt.xlabel('meanfun')
plt.ylabel('minfun')
plt.title('K-means Clustering for Male and Female Data')
plt.legend()
plt.show()
```

```python
# E4

from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix
import numpy as np
import pandas as pd

# Load the dataset
data = pd.read_csv("data_files/voice.csv")

# Select the columns
features = data[['meanfun', 'minfun']]
gender = data['label']

# Separate gender
male_data = features[gender == 'male']
female_data = features[gender == 'female']

# Fit Kmeans models to the data
kmeans_male = KMeans(n_clusters=1, random_state=13)
kmeans_male.fit(male_data)
kmeans_female = KMeans(n_clusters=1, random_state=13)
kmeans_female.fit(female_data)

# Predict the future data
predictions_male = kmeans_male.predict(male_data)
predictions_female = kmeans_female.predict(female_data)

# Combine the predictions for both male and female data
all_predictions = np.concatenate((predictions_male, predictions_female))

# Combine the true labels for both male and female data
all_true_labels = np.concatenate((np.zeros(len(predictions_male)), np.ones(len(predictions_female))))

# Convert predictions and true labels to pandas series
all_predictions_series = pd.Series(all_predictions)
all_true_labels_series = pd.Series(all_true_labels)

# Define a function to assign cluster labels based on majority vote
def assign_cluster_labels(predictions, true_labels):
    cluster_labels = {}
    unique_clusters = set(predictions)
    for cluster in unique_clusters:
        cluster_indices = (predictions == cluster)
        majority_label = true_labels[cluster_indices].mode().iloc[0]  # get common label
        cluster_labels[cluster] = majority_label
    return cluster_labels

# Assign cluster labels based on majority vote
cluster_labels = assign_cluster_labels(all_predictions_series, all_true_labels_series)

# Construct confusion matrix
conf_matrix = confusion_matrix(all_true_labels_series, [cluster_labels[pred] for pred in all_predictions_series])

# Display matrix
print("Confusion Matrix:")
print(conf_matrix)
```
